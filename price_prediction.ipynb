```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate sample data
number_of_samples = 1000
area = 120 * np.random.rand(number_of_samples) + 30
price = area * 1300 + 17700
noise = 20000 * np.random.rand(number_of_samples) - 10000
noisy_price = price + noise 
```

```python
# Fit linear regression model
lm = LinearRegression()
lm.fit(area.reshape(-1, 1), noisy_price)
```

```python
# Print the model parameters
print("Bias (Intercept):", lm.intercept_)
print("Weight (Coefficient):", lm.coef_)
```

```python
# Plot the data
plt.figure(figsize=(20, 10))
plt.scatter(area, price, label='Stable Price', alpha=0.6)
plt.scatter(area, noisy_price, label='Noisy Price', alpha=0.6)
plt.plot(area, lm.predict(area.reshape(-1, 1)), color='red', label='Fitted Line')

# Enhance plot aesthetics
plt.xlabel("Area (sq meters)", fontsize=14)
plt.ylabel("Price ($)", fontsize=14)
plt.title("Area vs. Price Relationship", fontsize=18)
plt.legend()
plt.grid(True)
plt.show()
```
